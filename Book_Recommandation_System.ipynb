```python
ratings = pd.read_csv(r"D:\Dataset\Book_Recommandation_System\BX-Book-Ratings.csv", sep=";", on_bad_lines="skip", encoding="latin-1", dtype={3: str})
```

```python
ratings.rename(columns={"User-ID":"user_id", "Book-Rating":"rating"},inplace=True)
```

```python
ratings['user_id'].value_counts()
```

```python
valid_users = ratings['user_id'].value_counts()>200
```

```python
valid_users[valid_users].shape
```

```python
rating_with_books = ratings.merge(books,on = "ISBN")
```

```python
rating_with_books.head()
```

```python
number_rating = rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = rating_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
